```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/datasets/Viral_Social_Media_Trends[1].csv")

print("initial shape:", df.shape)
print(df.head())
```

```python
# which posts generate the most engagement?

# DEFINE ENGAGEMENT METRICS
df['Engagement_Score']= df['Likes'] + df['Comments'] + df['Shares']

top10_posts= df.sort_values(by='Engagement_Score', ascending=False).head(10)
print(top10_posts[['Post_ID','Engagement_Score', 'Likes', 'Shares', 'Comments', 'Platform']])
```

```python
# VISUALIZE TOP 10 POSTS BY ENGAGEMENT SCORES

import matplotlib.pyplot as plt

top10_posts= df.sort_values('Engagement_Score', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10_posts['Post_ID'], top10_posts['Engagement_Score'], color='skyblue')
plt.xlabel('Engagement Score')
plt.ylabel('Post ID')
plt.title('Top 10 Posts by Engagement')
plt.gca().invert_yaxis
plt.show()
```

```python
# Which type of content drives engagement ?

hash_eng = df.groupby('Hashtag')['Engagement_Score'].mean().sort_values(ascending=False).head(10)
print(hash_eng)
```
